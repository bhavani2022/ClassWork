```python
def get_seconds(time_str):
    #print('Time in hh:mm:ss:', time_str)
    # split in hh, mm, ss
    hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
    
```

```python
avg_response_sec=[]
for t in df1['Average Response Time']:
    avg_response_sec.append(get_seconds(t))
avg_response_sec
df1['avg_response_sec'] = avg_response_sec
```

```python
df1.head()
```

```python
df1.groupby(['Agent Name','date_week'])['avg_response_sec'].mean()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['week'] = df['Date'].dt.isocalendar().week
```

```python
Total_duration_secs=[]
for t in df['Duration']:
    Total_duration_secs.append(get_seconds(t))
Total_duration_secs
df['Total_duration_secs'] = Total_duration_secs
```

```python
df['hours']=df['Total_duration_secs']//3600
```
