```python
"""q16 : Write a fun which will take input as a dict and give me out as a list of all the values even in case of 2 level 
nesting it should work . """
def diction():
#    dictn={"key1":"Bhavani", "key2":"gopal"}
    dictn={ 'dictA': {'key_1': 'value_1'},
            'dictB': {'key_2': 'value_2'}}
    l1=[]    
    l2=[]
    for k,v in dictn.items():
            l1.append(k) 
            l1.append(v)   
    for x in l1:
        if type(x) == dict:
            for m,n in x.items():
                l2.append(m) 
                l2.append(n) 
            
    print(l1+l2)
diction()
```
